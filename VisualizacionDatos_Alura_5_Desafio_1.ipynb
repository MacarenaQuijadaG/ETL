## **Desafio 5** creando una animación para comparar diferentes datos

```python
import pandas as pd
df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.set_index('Pais',inplace=True)
df
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
argentina = df.loc['Argentina',anos]
argentina
```

```python
arg_dict = {'Año':argentina.index.tolist(),
            'Inmigrantes':argentina.values.tolist()}
arg_dict
```

```python
brasil = df.loc['Brasil',anos]
brasil
```

```python
bra_dict = {'Año':brasil.index.tolist(),
            'Inmigrantes':brasil.values.tolist()}
bra_dict
```

```python
datos_brasil = pd.DataFrame(bra_dict)
datos_brasil.tail()
```

```python
datos_argentina = pd.DataFrame(arg_dict)
datos_argentina.tail()
```

```python
import plotly.graph_objs as go
```

```python
# Crear figura vacía
fig = go.Figure()

# Agregar las líneas iniciales con al menos 2 puntos (para que la línea se vea)
fig.add_trace(
    go.Scatter(
        x=datos_brasil['Año'].iloc[:2],
        y=datos_brasil['Inmigrantes'].iloc[:2],
        mode='lines+markers',
        name='Inmigrantes de Brasil',
        line=dict(width=4)
    )
)

fig.add_trace(
    go.Scatter(
        x=datos_argentina['Año'].iloc[:2],
        y=datos_argentina['Inmigrantes'].iloc[:2],
        mode='lines+markers',
        name='Inmigrantes de Argentina',
        line=dict(width=4)
    )
)

# Configurar diseño y ejes
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Brasil y Argentina a Canadá en el período de 1980 a 2013</b>',
        x=0.1,
        font=dict(size=18)
    ),
    xaxis=dict(
        range=[1980, 2013],
        autorange=False,
        title='<b>Año</b>',
        type='linear',  # asegúrate que es numérico lineal
        dtick=5  # marcas en el eje cada 5 años (opcional)
    ),
    yaxis=dict(
        range=[0, max(datos_brasil['Inmigrantes'].max(), datos_argentina['Inmigrantes'].max()) + 500],
        autorange=False,
        title='<b>Número de inmigrantes</b>'
    ),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200,
    height=600
)

# Crear frames de animación
frames = []
for i in range(2, len(datos_brasil)):
    frame_data = [
        go.Scatter(
            x=datos_brasil['Año'].iloc[:i+1],
            y=datos_brasil['Inmigrantes'].iloc[:i+1],
            mode='lines+markers'
        ),
        go.Scatter(
            x=datos_argentina['Año'].iloc[:i+1],
            y=datos_argentina['Inmigrantes'].iloc[:i+1],
            mode='lines+markers'
        )
    ]
    frames.append(go.Frame(data=frame_data))

fig.frames = frames

fig.show()
```
